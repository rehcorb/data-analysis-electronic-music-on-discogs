```python
# 1. Top 10 Artists
top_artists = analysis_df_cleaned['artist'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_artists.index, y=top_artists.values, palette='magma')
plt.title('Top 10 Artists in Electronic Music')
plt.xlabel('Artist')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Top 10 Styles
plt.figure(figsize=(10, 6))
genre_counts = analysis_df_cleaned['styles'].value_counts().head(10)  # Top 10 styles
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.title('Top 10 Styles in Electronic Music')
plt.xlabel('Styles')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Relationship between the number of "have" and "want"
plt.figure(figsize=(10, 6))
sns.scatterplot(x=analysis_df_cleaned['have'], y=analysis_df_cleaned['want'], alpha=0.6, color='mediumseagreen')
plt.title('Relationship between Number of "Have" and "Want"')
plt.xlabel('Number of Have')
plt.ylabel('Number of Want')
plt.show()
```

```python
# 4. Popularity of records based on ratings
plt.figure(figsize=(10, 6))
sns.scatterplot(x=analysis_df_cleaned['num_ratings'], y=analysis_df_cleaned['average_rating'], alpha=0.6, color='salmon')
plt.title('Average Rating vs. Number of Ratings')
plt.xlabel('Number of Ratings')
plt.ylabel('Average Rating')
plt.show()
```

```python
# 5. Price analysis based on the number of "wants"
plt.figure(figsize=(10, 6))
sns.scatterplot(x=analysis_df_cleaned['median_price'], y=analysis_df_cleaned['want'], alpha=0.6, color='gold')
plt.title('Median Price vs. Number of Wants')
plt.xlabel('Median Price')
plt.ylabel('Number of Wants')
plt.show()
```

```python
# 6. Price vs. Number of Buyers
plt.figure(figsize=(10, 6))
sns.scatterplot(x=analysis_df_cleaned['median_price'], y=analysis_df_cleaned['have'], alpha=0.6, color='royalblue')
plt.title('Median Price vs. Number of Buyers ("Have")')
plt.xlabel('Median Price')
plt.ylabel('Number of Buyers ("Have")')
plt.show()
```

```python
# 7. Histogram of Median Price
plt.figure(figsize=(10, 6))
sns.histplot(analysis_df_cleaned['median_price'], bins=30, kde=True, color='purple')
plt.title('Distribution of Median Prices')
plt.xlabel('Median Price')
plt.ylabel('Frequency')
plt.show()
```
